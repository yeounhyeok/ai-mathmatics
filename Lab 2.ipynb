### 실습 1: 손문제 1·3 검증 (7분)

```python
# Verify Problem 1
A = np.array([[1, -1], [2, 0]])
x = np.array([3.0, 1.0])
e1 = np.array([1.0, 0.0])
e2 = np.array([0.0, 1.0])

print('Ax =', A @ x)
print('Ae1 =', A @ e1, '← 1st column of A?', np.allclose(A @ e1, A[:, 0]))
print('Ae2 =', A @ e2, '← 2nd column of A?', np.allclose(A @ e2, A[:, 1]))
print('3*Ae1 + 1*Ae2 =', 3*(A@e1) + 1*(A@e2), '== Ax?', np.allclose(3*(A@e1)+1*(A@e2), A@x))

# Verify Problem 3
X_p3 = np.array([[2.,1.],[0.,3.],[4.,-1.]])
w_p3 = np.array([2., -1.])
b_p3 = 5.

ones = np.ones((3,1))
X_b = np.hstack([ones, X_p3])
w_b = np.concatenate([[b_p3], w_p3])

print('\ny_hat (X@w+b)    :', X_p3 @ w_p3 + b_p3)
print('y_hat (bias trick):', X_b @ w_b)
```
