## 🐛 버그 잡기 팀활동 (20분)

영상에서 제시된 버그입니다. 팀끼리 논의하며 해결하세요.

**버그 유형:** gradient 행렬 곱 순서 오류 — 오류 메시지 없이 잘못된 방향으로 학습

```python
# 🐛 Bug Hunt: GD runs but loss does not decrease
np.random.seed(5)
m_b, n_b = 30, 3
X_bug = np.hstack([np.ones((m_b, 1)), np.random.randn(m_b, n_b-1)])
y_bug = np.random.randn(m_b)

w_bug = np.zeros(n_b)
losses_bug = []

for _ in range(300):
    residual = X_bug @ w_bug - y_bug
    losses_bug.append(np.mean(residual**2))
    grad = (2/m_b) * X_bug.T @ residual    # 🐛 find the error
    w_bug = w_bug - 0.01 * grad
    

plt.figure(figsize=(6, 3))
plt.plot(losses_bug, color='tomato', lw=2)
plt.xlabel('Iteration'); plt.ylabel('MSE')
plt.title('Loss curve (buggy)'); plt.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()

print('Shape check:')
print('  X_bug.shape    :', X_bug.shape)
print('  residual.shape :', (np.random.randn(m_b)).shape)
#print('  X_bug @ residual  → shape?', (X_bug @ np.random.randn(m_b)).shape)
print('  X_bug.T @ residual → shape?', (X_bug.T @ np.random.randn(m_b)).shape)
print('  w_bug.shape    :', w_bug.shape)

print(losses_bug[:1])  # 처음 5개 값
print(losses_bug[-1:]) # 마지막 5개 값
```

**팀 토론 질문:**

1. `X_bug @ residual`과 `X_bug.T @ residual`의 shape을 각각 손으로 계산해보세요.
2. gradient의 shape은 $\mathbf{w}$의 shape과 같아야 합니다. 어느 쪽이 맞나요?
3. 이 버그가 왜 오류 없이 실행되는지 설명하세요. (힌트: $(30, 3) \times (30,)$ — 가능한가?)
4. 수학적으로 `X @ r`은 무엇을 계산하는 것이고, `X.T @ r`은 무엇을 계산하는 건가요?

```python
# ---- Fixed version (팀이 완성) ----
w_fix = np.zeros(n_b)
losses_fix = []

for _ in range(300):
    residual = X_bug @ w_fix - y_bug
    grad = (2/m_b) * X_bug.T @ residual    # fix: X or X.T?
    w_fix = w_fix - 0.01 * grad
    losses_fix.append(np.mean(residual**2))

plt.figure(figsize=(6, 3))
plt.plot(losses_bug, color='tomato', lw=2, label='Buggy')
plt.plot(losses_fix, color='steelblue', lw=2, label='Fixed')
plt.xlabel('Iteration'); plt.ylabel('MSE')
plt.title('Buggy vs Fixed'); plt.legend(); plt.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()

# 버그 원인:    행렬곱 차원이 맞지 않음
# 수학적 설명:  X는 y차원으로 변환시키는 행렬이고 X.T는 gradient 차원으로 변환하는 행렬인데, 경사하강법의 경우 후자이다.
# 왜 에러 없이 실행됐나: 잘못된 식이라도 X의 shape이 우연히 맞아떨어지면 계산 자체는 될 수 있다.
# X.T가 gradient 차원으로 변환하는 행렬임을 인지한다면 논리적으로도 y차원의 잔차를 내적하였을때 가중치 벡터가 나옴을 알수 있었다.
```

## 💻 코딩 실습 (20분)

### 실습 1: 손문제 검증 (7분)

```python
# Verify hand calculations from Problems 1-3
X_hw = np.array([[1, 2, -1],
                 [1, 0,  3],
                 [1, 1,  1],
                 [1, 4, -2]], dtype=float)
w_hw = np.array([0.5, 1.0, -0.5])
y_hw = np.array([3.0, -1.0, 2.0, 5.0])

# Problem 1: predictions
y_hat = X_hw @ w_hw
print('y_hat (Problem 1):', y_hat)

# Problem 2: gradient
m_hw = len(y_hw)
r = y_hat - y_hw
grad = (2/m_hw) * X_hw.T @ r
print('residual r       :', r)
print('gradient         :', grad)
print('gradient shape   :', grad.shape, '== w shape?', grad.shape == w_hw.shape)

# Problem 3: one GD step
lr = 0.1
w_new = w_hw - lr * grad
print('\nw_new            :', w_new)
mse_before = np.mean(r**2)
mse_after  = np.mean((X_hw @ w_new - y_hw)**2)
print(f'MSE before: {mse_before:.4f}  after: {mse_after:.4f}  improved: {mse_after < mse_before}')
```

### 실습 2: 정규화 효과 확인 (13분)

```python
# Compare GD convergence: different feature scales, with vs without normalization
np.random.seed(1)
m_r, n_r = 60, 3

# Features on very different scales
X_raw = np.random.randn(m_r, n_r) * np.array([1.0, 20.0, 0.05])
y_r   = 2*X_raw[:,0] + 0.1*X_raw[:,1] - 5*X_raw[:,2] + np.random.randn(m_r)*0.3

def run_gd(X, y, lr, n_iter=500):
    m_d, n_d = X.shape
    w = np.zeros(n_d)
    losses = []
    for _ in range(n_iter):
        r_d = X @ w - y
        w = w - lr * (2/m_d) * X.T @ r_d
        losses.append(np.mean(r_d**2))
    return w, losses

# Without normalization
X_raw_b = np.hstack([np.ones((m_r,1)), X_raw])
_, losses_raw  = run_gd(X_raw_b, y_r, lr=0.001) # 직접 lr을 찾아보세요

# With normalization
X_norm   = (X_raw - X_raw.mean(axis=0)) / X_raw.std(axis=0)
X_norm_b = np.hstack([np.ones((m_r,1)), X_norm])
_, losses_norm = run_gd(X_norm_b, y_r, lr=0.1)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(losses_raw,  color='tomato',    lw=2, label='Raw features')
axes[0].plot(losses_norm, color='steelblue', lw=2, label='Normalized')
axes[0].set_xlabel('Iteration'); axes[0].set_ylabel('MSE')
axes[0].set_title('Convergence comparison')
axes[0].legend(); axes[0].grid(True, alpha=0.3)

# Loss contour for 2 of the weights (normalized case)
w_opt, _ = run_gd(X_norm_b, y_r, lr=0.1, n_iter=2000)
print('Converged weights (normalized):', w_opt.round(3))
print('Final MSE (normalized):', losses_norm[-1].round(4))
```
