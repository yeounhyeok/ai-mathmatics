# 인공지능수학 — 손풀이 과제
## 편미분과 행렬 gradient의 연결

---

- **제출 방법:** 풀이 과정이 모두 보이도록 사진 촬영 또는 스캔 후 LMS 제출
- **채점:** P/F — 풀이 과정이 적혀 있으면 P (답이 틀려도 과정이 있으면 P)
- **과제 목표:** 편미분을 하나씩 계산하는 것과 행렬 gradient $\frac{2}{m}\mathbf{X}^T\mathbf{r}$이 동일함을 손으로 확인합니다.

## 핵심 메시지

> 편미분을 $w_0, w_1, w_2$에 대해 각각 계산한 결과를 벡터로 쌓으면 $\frac{2}{m}\mathbf{X}^T\mathbf{r}$과 정확히 같습니다.
>
> 행렬 표기는 새로운 계산이 아니라 편미분 $m$개를 한 번에 쓰는 압축된 표현입니다.
>
> 이것이 5주차에서 "특성이 $n$개여도 코드 한 줄이 바뀌지 않는" 이유입니다.
