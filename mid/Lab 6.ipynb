### 문제 5: 종합 — BCE Gradient 유도 (10분)

아래 함수를 $w$에 대해 미분하세요. ($x, y$는 상수, $y \in \{0, 1\}$)

$$L(w) = -\left[y\ln\sigma(wx) + (1-y)\ln(1-\sigma(wx))\right]$$

이 함수는 **이진 교차 엔트로피(BCE)** 손실함수입니다. 8주차 로지스틱회귀의 핵심입니다.

**단계별 유도:**

1단계 — 미분 분배 (선형성):
$$\frac{dL}{dw} = -\left[y \cdot \frac{d}{dw}\ln\sigma(wx) + (1-y) \cdot \frac{d}{dw}\ln(1-\sigma(wx))\right]$$

2단계 — 문제 3의 결과 대입:  
$\frac{d}{dw}\ln\sigma(wx) = $ ______ (문제 3-a 결과)  
$\frac{d}{dw}\ln(1-\sigma(wx)) = $ ______ (문제 3-b 결과)

$$\frac{dL}{dw} = -\left[y \cdot (\quad) + (1-y) \cdot (\quad)\right]$$

3단계 — 정리:
$$= -\left[yx(1-\sigma(wx)) - (1-y)x\sigma(wx)\right]$$

$$= -x\left[y - y\sigma(wx) - \sigma(wx) + y\sigma(wx)\right]$$

$$= -x\left[y - \sigma(wx)\right]$$

$$\boxed{\frac{dL}{dw} = (\sigma(wx) - y)\cdot x}$$

**결과 해석:**

- $\sigma(wx) > y$ (예측이 너무 큼): gradient > 0 → $w$를 줄이는 방향

- 형태가 MSE gradient $(\hat{y}-y)\cdot x$와 **구조적으로 동일**합니다!

## 🐛 버그 잡기 팀활동 (15분)

영상에서 제시된 버그입니다. 팀끼리 논의하며 해결하세요.

**버그 유형:** 시그모이드 미분 부호 오류 — 실행은 되지만 모든 값이 틀림

```python
# 🐛 Bug Hunt
def sigmoid_grad_buggy(x):
    s = sigmoid(x)
    return s * (1 + s)     # 🐛

x_vals = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])
print(f"{'x':>6}  {'buggy':>10}  {'numerical':>10}  {'match':>6}")
for xv in x_vals:
    buggy = sigmoid_grad_buggy(xv)
    num   = (sigmoid(xv+eps) - sigmoid(xv-eps)) / (2*eps)
    print(f'{xv:>6.1f}  {buggy:>10.6f}  {num:>10.6f}  {str(np.isclose(buggy,num)):>6}')
```

**팀 토론 질문:**

1. `s * (1+s)`와 `s * (1-s)` 중 어느 것이 $\sigma(x)(1-\sigma(x))$의 올바른 구현인가요?
2. `1+s`를 쓰면 수학적으로 어떤 함수를 계산하게 되나요?
3. $x=0$에서 buggy 값과 정확한 값을 손으로 계산해서 비교해보세요.

```python
# ---- Fixed version ----
def sigmoid_grad_fixed(x):
    s = sigmoid(x)
    return s * (1 - s)          # fix

print(f"{'x':>6}  {'fixed':>10}  {'numerical':>10}  {'match':>6}")
for xv in x_vals:
    fixed = sigmoid_grad_fixed(xv)
    num   = (sigmoid(xv+eps) - sigmoid(xv-eps)) / (2*eps)
    print(f'{xv:>6.1f}  {fixed:>10.6f}  {num:>10.6f}  {str(np.isclose(fixed,num)):>6}')

print()
print('버그 원인:   시그모이드함수 미분 틀림')
print('수학적 설명: 미분결과는 상기 함수 리턴값과 같다')
```
